```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
def f_0(x):
    if (x == 1.0) :
        return 5.0;
    else :
        return 0.0;
    
def f_1(x):
    return (x+1);

def f_2(x):
    return 2.5*x*x;

def R_mt(f, mu, alpha):
    R_max = sys.float_info.min;
    X = np.linspace(-1.0, 1.0, 100);
    for x in X :
        R = f(x) + mu * (np.sqrt(1 - x * x) - 1);
        if R > R_max:
            R_max = R
    return R_max

def R_mx(f, mu, alpha):
    R_max = sys.float_info.min;
    X = np.linspace(-1.0, 1.0, 100);
    for x in X :
        R = f(x) - alpha + (np.sqrt(1 - x * x) - 1);
        if R > R_max:
            R_max = R
    return R_max
```

```python
def f_2(x):
    return 5*x*x;

# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

x_min = 0.0;
x_max = 6.0;

N = 100
Mu = np.linspace(x_min, x_max, N, endpoint=True)

alpha = 1.0;
R_max = sys.float_info.min;
R_min = sys.float_info.max;
R_mutator = [];
R_mixed = [];
for mu in Mu:
    R1 = R_mt(f_2, mu, alpha);
    R2 = R_mx(f_2, mu, alpha);
    R_mutator.append(R1);
    R_mixed.append(R2);
    R_max = max(R_max, R1, R2);
    R_min = min(R_min, R1, R2);
    

# Set x limits
plt.xlim(x_min, x_max)

# Set x ticks
plt.xticks(np.linspace(x_min, x_max, 11, endpoint=True))

# Set y limits
y_min = R_min;
y_max = R_max;
plt.ylim(y_min, y_max)

# Set y ticks
plt.yticks(np.linspace(y_min, y_max, 5, endpoint=True))

# Numerics data
MuN = [ 1.0, 2.0, 3.0, 4.0 ];
RN = [ 4.05025, 3.201, 3.05025, 3.05025 ];
sN = [ 0.899499, 0.798999, 0.779547, 0.779227 ];
qN = [ 1.0, 1.0, 0.999987, 0.999463 ]

plt.plot(Mu, R_mutator, 'b-', label='$R_{mut}$', linewidth=1)
plt.plot(Mu, R_mixed, 'g-', label='$R_{mix}$', linewidth=1)
plt.plot(MuN, RN, 'ro', label='$R_{num}$')
plt.ylabel('$R$')
plt.xlabel('$\mu$')
plt.legend(loc='best')

# Save figure using 72 dots per inch
plt.savefig("mutator.png", dpi=72)

# Show result on screen
plt.show()
```
